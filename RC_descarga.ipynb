```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.arange(0, 20, 0.0001)
```

```python
E = 5

R = 1000

C = 0.000001

Q = 5

B = E*C

tau = R*C
```

```python
B
```

```python
def dq(T):
    return B * np.exp(-T/tau)
```

```python
t_dq = dq(T)
```

```python
tau_dq = dq(tau)
tau_dq
```

```python
dq(B)
```

```python
plt.plot(T, t_dq, c='y')
plt.xlabel('$tiempo (s)$')
plt.ylabel('Descarga ($C$)')
plt.title('$q(t)=Q e^{-t/RC}$ \n Descarga de un Capacitor en funci√≥n del tiempo')
plt.vlines(tau, ymin=0, ymax=tau_dq, color='grey', linestyles='--')
plt.hlines(tau_dq, xmin=0, xmax=tau, color='grey', linestyles='--')
plt.ylim(0)
plt.xlim(0, 0.01)
plt.text(0.002, 1.5e-6, '$t=RC$')
plt.text(0.0005, 2e-6, '$1.8e-6 = \epsilon C  $')
#plt.text(0.04, 3e-6, '$Q$', fontsize=12, color='red')
#plt.savefig('descarga_3.png')
plt.show()
```
