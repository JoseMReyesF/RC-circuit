```python
import numpy as np
import matplotlib.pyplot as plt
```

### Time

```python
T = np.arange(0, 20, 0.001)
```

```python
T
```

### Constantes

```python

E = 5

R = 12000 

C = 0.000001

Q = 5

B = E*C
tau = R*C
```

### Carga

$$q(t) = \epsilon * C(1-e^{-t/\tau})$$

```python
def q(T):
    return B*(1-np.exp(-T/tau))
```

```python
ch = q(T)
```

```python
ch
```

```python
ec = q(tau)
```

```python
tau
```

```python
ec
```

```python
plt.plot(T, ch, color='purple')
plt.xlabel('$tiempo (s)$')
plt.ylabel('Carga ($C$)')
plt.title('$q(t)=\epsilon C(1-e^{-t/RC})$ \n Simulacin de la carga de un \n Capacitor en funcin del tiempo')
plt.axhline(B, color='red', linestyle='--')
plt.vlines(tau, ymin=0, ymax=ec, color='grey', linestyles='--')
plt.hlines(ec, xmin=0, xmax=tau, color='grey', linestyles='--')
plt.vlines(0.049, ymin=0, ymax=5e-6, color = 'green', label='Tiempo de carga real')
plt.ylim(0)
plt.xlim(0, .125)
plt.text(0.015, 3e-6, '$t=RC$')
plt.text(0.005, 3.3e-6, '$\epsilon C$')
plt.text(0.055, 4.3e-6, '$t = 0.049$ \n Tiempo de carga \n Real')
plt.legend()
#plt.text(10, 1, '$\epsilon C$', fontsize=12, color='red')
plt.savefig('carga_3.png')
plt.show()
```

### Corriente

```python
def i(T):
    return B * np.exp(-T/tau)
```

```python
ip  = i(T)
```

```python
ip
```

```python
ir = i(tau)
```

```python
ir
```

```python
plt.plot(T, ip, color='brown')
plt.xlabel('tiempo ($s$)')
plt.ylabel('Corriente ($A$)')
plt.axhline(1, color='red', linestyle='--')
plt.vlines(tau, ymin=0, ymax=ir, color='grey', linestyles='--')
plt.hlines(ir, xmin=0, xmax=tau, color='grey', linestyles='--')
plt.xlim(0,)
plt.ylim(0)
plt.title('$dQ/dt= i = \epsilon/R e^{-t/RC}$ \n Corriente en funcin del tiempo \n en circuito con capacitor en carga')
#plt.text(14, 0.1, '$t=RC$')
#plt.text(2.2, 0.4, '$0.368 \epsilon / R$')
#plt.text(10, 1.01, '$\epsilon/ R$', fontsize=12, color='red')
#plt.savefig('descarga1_1.png')
plt.show()
```

```python

E = 5

R = 12000 

C = 0.000001

Q = 5

B = E*C
tau = R*C
```

```python
def dq(T):
    return B * np.exp(-T/tau)
```

```python
dq = dq(T)
```

```python
dq
```

```python
plt.plot(T, dq, c='y')
plt.xlabel('$tiempo (s)$')
plt.ylabel('Descarga ($C$)')
plt.title('$q(t)=Q e^{-t/RC}$ \n Descarga de un Capacitor en funcin del tiempo')
plt.vlines(tau, ymin=0, ymax=ir, color='grey', linestyles='--')
plt.hlines(ir, xmin=0, xmax=tau, color='grey', linestyles='--')
plt.ylim(0)
plt.xlim(0, 0.12)
plt.text(0.02, 1.5e-6, '$t=RC$')
#plt.text(2.2, 0.4, '$0.368 \epsilon C$')
#plt.text(1, 1, '$Q$', fontsize=12, color='red')
#plt.savefig('descarga_2.png')
plt.show()
```

```python
def id(T):
    return -1/tau * np.exp(-T/tau)
```

```python
id = id(T)
```

```python
plt.plot(T, id, color='blue')
plt.xlabel('tiempo ($s$)')
plt.ylabel('Corriente ($A$)')
plt.xlim(0)
plt.ylim(-0.083)
plt.axhline(0, color='black', linestyle='--')
plt.vlines(tau, ymin=-0.083, ymax=-0.0306, color='grey', linestyles='--')
plt.hlines(-0.0306, xmin=0, xmax=12, color='grey', linestyles='--')
plt.text(13, -0.035, '$t=RC$')
plt.text(1, -0.025, '$-0.368  Q/RC$')
plt.title('$dq/dt=i=-Q/RC e^{-(t/RC)}$\n Corriente en funcin \n del tiempo en circuito \n con capacitor en descarga')
plt.savefig('corriente_descarga_1.png')
plt.show()
```
